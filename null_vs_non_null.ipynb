# crash data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('only_crash.csv')

# Ensure plots display in Colab
%matplotlib inline

# Loop through columns
for col in df.columns:
    total = len(df)
    null_count = df[col].isnull().sum()
    non_null_count = total - null_count

    # Skip if no missing values
    if null_count == 0:
        continue

    # Plot pie chart
    plt.figure(figsize=(4, 4))
    plt.pie(
        [non_null_count, null_count],
        labels=['Non-null', 'Null'],
        autopct='%1.1f%%',
        colors=['#4CAF50', '#F44336'],
        startangle=90
    )
    plt.title(f'Null vs Non-null in "{col}"')
    plt.tight_layout()
    plt.show()
```
